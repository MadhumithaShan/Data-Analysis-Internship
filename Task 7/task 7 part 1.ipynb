```python
import pandas as pd
from sklearn.datasets import load_iris


iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

df_iris.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


features = df_iris.drop('target', axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)


pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)


df_iris['PCA1'] = pca_result[:, 0]
df_iris['PCA2'] = pca_result[:, 1]
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8,6))
for target in df_iris['target'].unique():
    plt.scatter(df_iris[df_iris['target'] == target]['PCA1'],
                df_iris[df_iris['target'] == target]['PCA2'],
                label=iris.target_names[target])
plt.title('PCA of Iris Dataset')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend()
plt.grid(True)
plt.show()
```
